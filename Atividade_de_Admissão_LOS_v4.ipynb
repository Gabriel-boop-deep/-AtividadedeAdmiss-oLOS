3.B - Corte as matrizes lbl e img em várias matrizes 64x64 e armazene essas matrizes em duas listas separadas, descartando sobras nas bordas. ✂↔📂

```python
import numpy as np

# dividir a matriz em 4 partes
num_sections = 4

if num_sections > 0:
    img_arr_split = np.array_split(img_arr, num_sections)
else:
    print("O número de seções deve ser maior que zero.")
```

```python
import numpy as np

# tamanho do corte das matrizes
CUT_SIZE = 64

# conversão da matriz lbl para booleanos
lbl_bool = lbl_arr >= 0.99

# listas para armazenar as matrizes 64x64 de óleo e não óleo
img_64_oleo = []
lbl_64_oleo = []
img_64_nao_oleo = []
lbl_64_nao_oleo = []

# define o tamanho da seção
section_size = 64

# calcula o número máximo de seções
max_sections_x = img_arr.shape[1] // section_size
max_sections_y = img_arr.shape[0] // section_size

# verifica se há pelo menos uma seção
if max_sections_x > 0 and max_sections_y > 0:
    # faz o corte das matrizes em seções
    lbl_sections = [lbl_arr[y * section_size : (y+1) * section_size, x * section_size : (x+1) * section_size] for y in range(max_sections_y) for x in range(max_sections_x)]
    img_sections = [img_arr[y * section_size : (y+1) * section_size, x * section_size : (x+1) * section_size] for y in range(max_sections_y) for x in range(max_sections_x)]


# cortando as matrizes em partes de 64x64 e armazenando em suas respectivas listas
for i, img_part in enumerate(np.array_split(img_arr, img_arr.shape[0] // CUT_SIZE)):
    for j, img_subpart in enumerate(np.array_split(img_part, img_part.shape[1] // CUT_SIZE, axis=1)):
        # verifica se a matriz de 64x64 contém óleo ou não
        if np.any(lbl_bool[i*CUT_SIZE:(i+1)*CUT_SIZE, j*CUT_SIZE:(j+1)*CUT_SIZE]):
            img_64_oleo.append(img_subpart)
            lbl_64_oleo.append(lbl_bool[i*CUT_SIZE:(i+1)*CUT_SIZE, j*CUT_SIZE:(j+1)*CUT_SIZE])
        else:
            img_64_nao_oleo.append(img_subpart)
            lbl_64_nao_oleo.append(lbl_bool[i*CUT_SIZE:(i+1)*CUT_SIZE, j*CUT_SIZE:(j+1)*CUT_SIZE])

# convertendo as listas de matrizes em arrays numpy
img_64_oleo = np.array(img_64_oleo)
lbl_64_oleo = np.array(lbl_64_oleo)
img_64_nao_oleo = np.array(img_64_nao_oleo)
lbl_64_nao_oleo = np.array(lbl_64_nao_oleo)
```

3.C - Considerando as listas lbl e img, monte dois arrays lbl e img balanceados com 50% óleo e 50% não óleo. Use os valores booleanos da lbl para filtrar. ⚖

```python
# Contadores para manter o balanço entre as classes
count_oil = 0
count_non_oil = 0

# Iterando sobre as listas de imagens e rótulos
for img, lbl_bool in zip(lbl_arr, lbl_bool):
    
    # Filtrando imagens e rótulos com base no valor booleano de lbl_bool
    if lbl_bool:
        if count_oil < len(img_arr)//2:
            # Adicionando a imagem e o rótulo à lista de óleo
            img_arr = np.vstack((img_arr, np.expand_dims(img, axis=0)))
            lbl_bool_arr = np.vstack((lbl_bool_arr, np.expand_dims(lbl_bool, axis=0)))
            count_oil += 1
    else:
        if count_non_oil < len(img_arr)//2:
            # Adicionando a imagem e o rótulo à lista de não-óleo
            img_arr = np.vstack((img_arr, np.expand_dims(img, axis=0)))
            lbl_bool_arr = np.vstack((lbl_bool_arr, np.expand_dims(lbl_bool, axis=0)))
            count_non_oil += 1
    
    # Verificando se já temos imagens suficientes de ambas as classes
    if count_oil >= len(img_arr)//2 and count_non_oil >= len(img_arr)//2:
        break

# Removendo a primeira linha dos arrays vazios
img_arr = img_arr[1:]
lbl_bool_arr = lbl_bool_arr[1:]

# Embaralhando os dados
indices = np.random.permutation(len(img_arr))
img_arr = img_arr[indices]
lbl_bool_arr = lbl_bool_arr[indices]

# Salvando os arrays como arquivos NPZ
np.savez_compressed('Arquivos/img_arr.npz', img_arr)
np.savez_compressed('Arquivos/lbl_bool_arr.npz', lbl_bool_arr)
```
