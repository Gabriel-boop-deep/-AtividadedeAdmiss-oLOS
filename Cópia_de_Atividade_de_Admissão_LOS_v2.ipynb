3.A - Converta os valores flutuantes da matriz lbl para booleanos, sendo assim todo valor maior ou igual a 0,99 deve ser verdadeiro. üÖæ1Ô∏è‚É£

```python
#Podemos converter os valores flutuantes em booleanos dessa forma:
array_ibl_bool = array_ibl >= 0.99
#Conferindo
# array_ibl_bool
```

3.B - Corte as matrizes lbl e img em v√°rias matrizes 64x64 e armazene essas matrizes em duas listas separadas, descartando sobras nas bordas. ‚úÇ‚ÜîüìÇ

```python
# Fun√ß√£o para descartar sobras nas bordas, a fun√ß√£o pode ser utilizada mais para frente, se assim existir necessidade
# E dar para mudar os parametros facilmente, para estudarmos como o c√≥digo se comporta em diferentes cen√°rios
def discard_edges(arr, size):
    """
    Nesse primeiro momento vamos descartar as sobras nas bordas de um array multidimensional.

    Legenda dos Par√¢metros:
    ========================================================================================
    arr : numpy array
        O array a ser cortado.
    size : int
        O tamanho do corte.
    ========================================================================================
    Retorno:
    ========================================================================================
    Uma lista com as submatrizes cortadas.
    ========================================================================================
    """
    return [arr[i:i+size, j:j+size] for i in range(0, arr.shape[0], size) 
            for j in range(0, arr.shape[1], size) if i+size <= arr.shape[0] and j+size <= arr.shape[1]]

# Cortar as matrizes lbl e img em v√°rias matrizes 64x64
size = 64
lbl_list = discard_edges(array_ibl_bool, size)
img_list = discard_edges(array_img ,size)
```

3.C - Considerando as listas lbl e img, monte dois arrays lbl e img balanceados com 50% √≥leo e 50% n√£o √≥leo. Use os valores booleanos da lbl para filtrar. ‚öñ

```python
# criando listas vazias para armazenar as submatrizes 64x64
lbl_list = []
img_list = []

# definindo a propor√ß√£o desejada para cada classe
prop = 0.5

# obtendo os √≠ndices dos pixels rotulados como √≥leo e n√£o √≥leo
oil_idx = np.where(array_ibl_bool)
no_oil_idx = np.where(~array_ibl_bool)

# obtendo o n√∫mero de amostras necess√°rias para cada classe com base na propor√ß√£o pedida.
n_oil = int(len(oil_idx[0]) * prop)
n_no_oil = int(len(no_oil_idx[0]) * prop)

# selecionando amostras aleat√≥rias de cada classe
oil_samples = random.sample(range(len(oil_idx[0])), n_oil)
no_oil_samples = random.sample(range(len(no_oil_idx[0])), n_no_oil)

# adicionando as submatrizes correspondentes a cada amostra nas listas lbl_list e img_list e para isso iremos usar o loop for
for i in oil_samples:
    row = oil_idx[0][i]
    col = oil_idx[1][i]
    lbl_list.append(array_ibl[row-32:row+32, col-32:col+32])
    img_list.append(array_img[row-32:row+32, col-32:col+32])

for i in no_oil_samples:
    row = no_oil_idx[0][i]
    col = no_oil_idx[1][i]
    lbl_list.append(array_ibl[row-32:row+32, col-32:col+32])
    img_list.append(array_img[row-32:row+32, col-32:col+32])

# convertendo as listas em arrays(Assim, os arrays ficam com dimens√µes irregulares para a configura√ß√£o que estamos concluindo)
# lbl_array = np.array(lbl_list)
# img_array = np.array(img_list)
lbl_array = np.array(lbl_list, dtype=object)
img_array = np.array(img_list, dtype=object)
```
