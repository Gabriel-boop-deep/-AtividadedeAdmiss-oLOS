**1.B** - Converta os dois arquivos PNG carregados para o tipo array, afinal ja discutimos que imagens são matrizes. 🎴↔🔢<br>

3.B - Corte as matrizes lbl e img em várias matrizes 64x64 e armazene essas matrizes em duas listas separadas, descartando sobras nas bordas. ✂↔📂

```python
# Função para descartar sobras nas bordas, a função pode ser utilizada mais para frente, se assim existir necessidade
# E dar para mudar os parametros facilmente, para estudarmos como o código se comporta em diferentes cenários
def discard_edges(arr, size):
    """
    Nesse primeiro momento vamos descartar as sobras nas bordas de um array multidimensional.

    Legenda dos Parâmetros:
    ========================================================================================
    arr : numpy array
        O array a ser cortado.
    size : int
        O tamanho do corte.
    ========================================================================================
    Retorno:
    ========================================================================================
    Uma lista com as submatrizes cortadas.
    ========================================================================================
    """
    return [arr[i:i+size, j:j+size] for i in range(0, arr.shape[0], size) 
            for j in range(0, arr.shape[1], size) if i+size <= arr.shape[0] and j+size <= arr.shape[1]]

# Cortar as matrizes lbl e img em várias matrizes 64x64
size = 64
lbl_list = discard_edges(array_ibl_bool, size)
img_list = discard_edges(array_img ,size)
```

3.C - Considerando as listas lbl e img, monte dois arrays lbl e img balanceados com 50% óleo e 50% não óleo. Use os valores booleanos da lbl para filtrar. ⚖

```python
# criando listas vazias para armazenar as submatrizes 64x64
lbl_list = []
img_list = []

# definindo a proporção desejada para cada classe
prop = 0.5

# obtendo os índices dos pixels rotulados como óleo e não óleo
oil_idx = np.where(array_ibl_bool)
no_oil_idx = np.where(~array_ibl_bool)

# obtendo o número de amostras necessárias para cada classe com base na proporção pedida.
n_oil = int(len(oil_idx[0]) * prop)
n_no_oil = int(len(no_oil_idx[0]) * prop)

# selecionando amostras aleatórias de cada classe
oil_samples = random.sample(range(len(oil_idx[0])), n_oil)
no_oil_samples = random.sample(range(len(no_oil_idx[0])), n_no_oil)

# adicionando as submatrizes correspondentes a cada amostra nas listas lbl_list e img_list e para isso iremos usar o loop for
for i in oil_samples:
    row = oil_idx[0][i]
    col = oil_idx[1][i]
    lbl_list.append(array_ibl[row-32:row+32, col-32:col+32])
    img_list.append(array_img[row-32:row+32, col-32:col+32])

for i in no_oil_samples:
    row = no_oil_idx[0][i]
    col = no_oil_idx[1][i]
    lbl_list.append(array_ibl[row-32:row+32, col-32:col+32])
    img_list.append(array_img[row-32:row+32, col-32:col+32])

# convertendo as listas em arrays(Assim, os arrays ficam com dimensões irregulares para a configuração que estamos concluindo)
# lbl_array = np.array(lbl_list)
# img_array = np.array(img_list)
lbl_array = np.array(lbl_list, dtype=object)
img_array = np.array(img_list, dtype=object)
```
